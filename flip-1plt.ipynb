```python
c10f2 = wb['C10-flip2']
```

```python
def move_avg(nums, window):
    l = len(nums)
    new_nums = []
    for i in range(window,l):
        new_nums.append(sum(nums[i-window:i])/window)
    return new_nums
```

```python
tran = [x.value for x in c10f2['p'] if x.value is not None]
rot = [x.value for x in c10f2['q'] if x.value is not None]
t = [x.value for x in c10f2['r'] if x.value is not None]
tran, rot, t = tran[1:], rot[1:], t[1:]
```

```python
w = 100
avg_tran = move_avg(tran, w)
avg_rot = move_avg(rot, w)
```

```python
trace0 = go.Scatter(
    x = t[w-1:],
    y = avg_tran,
    mode = 'lines',
    line = dict(color = 'green'),
    name = 'tran'
)
trace1 = go.Scatter(
    x = t[w-1:],
    y = avg_rot,
    mode = 'lines',
    line = dict(color = 'blue'),
    name = 'rot'
)
```
